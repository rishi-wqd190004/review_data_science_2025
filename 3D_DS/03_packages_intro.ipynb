### Point cloud

```python
file_data_path = '../datasets/3D_DS/sample.xyz'
pt_cld = np.loadtxt(file_data_path, skiprows=1, max_rows=1000000)
```

```python
pt_cld[:,2]
```

```python
pt_cld_np = pt_cld
mean_z = np.mean(pt_cld_np, axis=0)[2]
spatail_query = pt_cld_np[abs(pt_cld_np[:,2]-mean_z)<1]

xyz = spatail_query[:, :3]
rgb = spatail_query[:,3:]

ax = plt.axes(projection='3d')

ax.scatter(xyz[:, 0], xyz[:,1], xyz[:,2], c=rgb/255, s=0.01)

plt.show()
```
